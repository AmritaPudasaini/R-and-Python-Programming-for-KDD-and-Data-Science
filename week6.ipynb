## Loading the dataset

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('cereal.csv')

# Replace -1 with NaN (important)
df.replace(-1, np.nan, inplace=True)

# Convert numeric columns
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')

# Check data
df.head()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

train = df[df['potass'].notna()].copy()

X = train.drop(columns=['potass', 'name', 'mfr', 'type'])
y = train['potass']

# convert target to numeric (invalid → NaN)
y = pd.to_numeric(y, errors='coerce')

# remove invalid target rows
mask = y.notna()
X = X[mask]
y = y[mask]

X = X.apply(pd.to_numeric, errors='coerce')

imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

model = LinearRegression()
model.fit(X, y)

row = df[df['name'] == 'Almond Delight'].drop(columns=['potass', 'name', 'mfr', 'type'])
row = row.apply(pd.to_numeric, errors='coerce')
row = imputer.transform(row)

pred = model.predict(row)[0]

residuals = y - model.predict(X)
se = residuals.std()

print("Predicted potassium:", pred)
print("Standard error:", se)
```

```python
row2 = df[df['name'].str.contains('Cream of Wheat', case=False, na=False)] \
        .drop(columns=['potass', 'name', 'mfr', 'type'])

row2 = row2.apply(pd.to_numeric, errors='coerce')
row2 = imputer.transform(row2)

pred2 = model.predict(row2)[0]

print("Q2 - Predicted potassium (Cream of Wheat):", pred2)
print("Q2 - Standard error:", se)
```
